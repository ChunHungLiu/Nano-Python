## 6. 線性迴歸

### 6-1 做個假的資料準備迴歸一下

我們來一條線, 比如說

$$f(x) = 1.2x + 0.8$$

準備好個 50 個點

````python
x = linspace(0, 5, 50)
````

我們可以畫出來看看。

````python
scatter(x, 1.2*x + 0.8)
plot(x, 1.2*x+0.8, 'b')
````

這要學嗎!! 真實世界不可能這樣的。

````python
y = 1.2*x + 0.8 + randn(50)
````

````python
scatter(x, y)
plot(x, 1.2*x+0.8, 'b')
````

像樣多了, 現在我們準備來學學 (當然假設不知藍色那條「正確答案」。)

### 6.2 用 SciPy 來迴歸

線性迴歸很多套件會做, 我們這裡介紹的方法好處是你只要預想好這些資料長得很像什麼函數, 我們都可以迴歸! 不一定要線性函數!

在我們線性的例子很容易。我們先做一個目標函數

$$f(x) = ax + b$$

再來算 a, b 這樣。

````python
from scipy.optimize import curve_fit
````

````python
def f(x, a, b):
    return a*x + b
````

我們「學習的資料」就是剛剛的 x, y。傳回的 popt 這裡就是最小平方法下得到的 a, b。

````python
popt, pcov = curve_fit(f, x, y)
````

````python
popt
````

````python
a, b = popt
````

````python
scatter(x,y)
plot(x,1.2*x + 0.8,'b',label="original")
plot(x, a*x + b, 'r',label="fit")
ax = gca()
ax.legend();
````

對了, 我們的函數寫好, 當然是可以預測的!

````python
def g(x):
    return a*x + b
````

````python
g(5.1)
````

````python
1.2*5.1 + 0.8
````

## 7. 精準學會的插值法

有時候我們會用「插值法」, 把一個函數精準學會。

比如我們想學:

$$f(x) = \sin(x) + 0.5x$$

### 7.1 生出訓練資料

````python
x = linspace(0, 10, 15)
y = sin(x) + 0.5 * x
````

````python
scatter(x,y)
````

為了畫圖, 我們要多取一些點, 叫 xx 好了。

````python
xx = linspace(0, 10, 100)
````

````python
scatter(x, y, s=50)
plot(xx, sin(xx) + 0.5*xx, 'b');
````

插值法的目標是全部命中!!

## 7.2 全部命中沒有想像中了不起

````python
def draw(a):
    scatter(x, y, s=50)
    xxx = linspace(0, 10, 1000)
    plot(xxx, 3.2*sin(a*xxx)+3, 'r');
````

````python
interact(draw, a=(1,100));
````

### 7.3 好像很了不起的 Lagarange 插值法

````python
from scipy.interpolate import lagrange
````

````python
L = lagrange(x,y)
````

````python
L(3)
````

````python
scatter(x, y, s=50)
plot(xx, L(xx), 'r');
````

### 7.4 Cubic Spline 插值法

````python
from scipy.interpolate import interp1d
````

````python
f = interp1d(x, y, kind='cubic')
````

````python
scatter(x, y, s=50)
plot(xx, sin(xx) + 0.5*xx, 'b') # 真正的
plot(xx, f(xx), 'r');
````

## 8. 假裝我們好像很懂機器學習之 SVM

在機器學習的分類 (其實也是一個學函數的過程) 大致有

* supervised learning (監督式學習)
* unsupervised learning (非監督式學習)

我們各舉一例, 假裝我們好像很懂。

### 8.1 SVM 支持向量機

SVM (Support Vector Machines) 是用直線、曲線、平面、曲面等等把各類資料分開的方式。

來個很簡單的分類。

[例子]
第一類: (-1, -1), (-2, -1)

第二類: (1, 1), (2, 1)

這樣告訴 Python 什麼樣的點是哪一類。

````python
X = array([[-1, -1], [-2, -1], [1, 1], [2, 1]])
Y = array([1, 1, 2, 2])
````

### 8.2 使用 scikit-learn (sklearn)

````python
from sklearn.svm import SVC
````

打開個 SVC!

````python
clf = SVC()
````

````python
clf.fit(X,Y)
````

來預測!

````python
clf.predict([[-0.8,-1]])
````

這告訴我們 (-0.8, -1) 會被歸到第 1 類!

我們先來看看原始資料。

````python
scatter(X[:,0], X[:,1],s=50,c=Y);
````

放一堆資料讓它分類看看!

````python
gd = array([[i,j] for i in arange(-2.5, 2.5, 0.2) 
 for j in arange(-2.5, 2.5, 0.2)])
````

````python
gdc = clf.predict(gd)
````

````python
scatter(gd[:,0], gd[:,1], s=50, c=gdc);
````

## 9. 假裝我們好像很懂機器學習之 K-Mean

### 9.1 K-Mean 是一種自動分類法

所謂非監督式學習就是「沒有標準解答」的學習。比如說們處理一些數據, 本來沒有分類, 讓電腦自動分類。

K-mean 是這種的方法。

````python
X = randn(50,2)
````

````python
scatter(X[:,0], X[:,1], s=55);
````

### 9.2 用 K-Mean 分三類!

````python
from sklearn.cluster import KMeans
````

````python
clf = KMeans(n_clusters=3)
````

````python
clf.fit(X)
````

偷看一下結果...

````python
clf.labels_
````

````python
scatter(X[:,0], X[:,1], s=55, c=clf.labels_)
````

### 9.3 不是訓練的點也可以分了

````python
gdc = clf.predict(gd)
scatter(gd[:,0], gd[:,1], s=50, c=gdc);
````
